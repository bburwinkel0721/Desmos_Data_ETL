```python
# Imports for the project
import pandas as pd
import numpy as np
import glob
```

```python
# Generalized file path so that we can just drop any csv into this folder for processing
file_path = glob.glob('Resources/*.csv')[0] 

# Read the data into a Pandas DataFrame and show it
assessment_data_df = pd.read_csv(file_path)
assessment_data_df.head()
```

```python
# Remove the second and thrid columns because they are not needed
assessment_data_df = assessment_data_df.drop(assessment_data_df.columns[[1,2]], axis=1)
```

```python

# Remove the last column because it is not needed
assessment_data_df = assessment_data_df.iloc[:,:-1]

# Show the dataframe
assessment_data_df.head()
```

```python
# Set the index to be the names of students
assessment_data_df.set_index(assessment_data_df.columns[0], inplace=True)
```

```python
# Display the dataframe to confirm changes
assessment_data_df
```

```python
# Replace blank answers with zeros
assessment_data_df.replace('-', 0, inplace=True)
```

```python
# Change all data types to integers for future calculations
assessment_data_df = assessment_data_df.astype(int)
```

```python
# Check data types to confirm changes
assessment_data_df.dtypes
```

```python
# Create a column for percentages on the assessment
assessment_data_df['Percentage'] = assessment_data_df.sum(axis=1)/(4*assessment_data_df.count(axis=1))

# Display the dataframe to confirm changes
assessment_data_df.head()
```

```python
# Add a column for the raw points to award students on a quiz
assessment_data_df['Raw Total'] = assessment_data_df['Percentage']*4

# Display the dataframe to confirm changes
assessment_data_df.head()
```

```python
# Calculate the mean of every column
avg_row = assessment_data_df.mean()

# Place the means of each column as a row at the bottom of the dataframe
assessment_data_df.loc['Mean'] = avg_row

# Display the dataframe to confirm changes
assessment_data_df
```

```python
# Round values in the dataframe to the nearest hundreth
assessment_data_df = assessment_data_df.round(2)

# Display the dataframe to confirm changes
assessment_data_df
```

```python
# Save the transformed dataframe as a new csv file 
assessment_data_df.to_csv('Resources/cleaned_assessment_data.csv', encoding='utf8', index=True)
```
